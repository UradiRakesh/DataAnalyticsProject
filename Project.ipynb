#  **3D** **Scatter** **Plot**

```python
import plotly.express as px
sctr = px.scatter_3d(statistics.head(25), x='Country',y='Research', z='Government Strategy',
              color='Region')
sctr.update_layout(title='Countries and their Government Strategies,Research in Top 25 Countries', autosize=True,
                  width=900, height=700,
                  margin=dict(l=65, r=50, b=65, t=90),
                 xaxis= dict(title= 'Country',ticklen= 100),
                 yaxis= dict(title= 'Research',ticklen= 100),
                template="plotly_dark",

                 font=dict(
        family="Times New Roman",
        size=15))
sctr.show()
```

# **Bar** **Plot**

```python

import plotly.graph_objects as px
import numpy

plot = px.Figure(data=[go.Bar(
    name='Development (TOP25)', #name of bar
    x=statistics.head(25)['Country'], #x-axis values
    y=statistics.head(25)['Development'] , #y-axis values
    marker = dict(color = 'blue',
                              line=dict(color='rgb(0,0,0)',width=1.0))#properties of bar like color,border
),
    go.Bar(
    name='Talent (TOP25)',
    x=statistics.head(25)['Country'],
    y=statistics.head(25)['Talent'],
        marker = dict(color = 'yellow',
                              line=dict(color='rgb(0,0,0)',width=1.0))
),
  go.Bar(
    name='Research (TOP25)',
    x=statistics.head(25)['Country'],
    y=statistics.head(25)['Research'],
      marker = dict(color = 'white',
                              line=dict(color='rgb(0,0,0)',width=1.0))
),
    go.Bar(
    name='Government Strategy (TOP25)',
    x=statistics.head(25)['Country'],
    y=statistics.head(25)['Government Strategy'],
    marker = dict(color = 'crimson',
                              line=dict(color='rgb(0,0,0)',width=1.0))
),
    go.Bar(
    name='Commercial (TOP25)',
    x=statistics.head(25)['Country'],
    y=statistics.head(25)['Commercial'],
    marker = dict(color = 'crimson',
                              line=dict(color='rgb(0,0,0)',width=1.0))
)
])


# Add dropdown
plot.update_layout(
    template="plotly_dark",
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="All",
                     method="update",
                     args=[{"visible": [True, True, True, True,True]},
                           {"title": "All"}]),
                dict(label="Development (TOP25)",
                     method="update",
                     args=[{"visible": [True, False, False, False,False]},
                           {"title": "Development (TOP25)",
                            }]),
                dict(label="Talent (TOP25)",
                     method="update",
                     args=[{"visible": [False, True, False, False,False]},
                           {"title": "Talent (TOP25)",
                            }]),
                dict(label="Research (TOP25)",
                     method="update",
                     args=[{"visible": [False, False, True, False,False]},
                           {"title": "Research (TOP25)",
                            }]),
                dict(label="Government Strategy (TOP25)",
                     method="update",
                     args=[{"visible": [False, False, False, True,False]},
                           {"title": "Government Strategy (TOP25)",
                            }]),
                dict(label="Commercial (TOP25)",
                     method="update",
                     args=[{"visible": [False, False, False, False,True]},
                           {"title": "Commercial (TOP25)",
                            }]),

            ]),
        )
    ])



```

# **WATERFALL** **CHART**

```python
import plotly.graph_objects as go

statistics = {'Talent': [1, 2, 3, 4, 5],
               'Total Score': [10, 20, 30, 40, 50]}

water = go.Figure(go.Waterfall(x=statistics['Talent'], y=statistics['Total Score'], base=0, increasing_marker_color='green', decreasing_marker_color='red'))

water.update_layout(title='Waterfall Chart',
                  xaxis_title='Talent',
                  yaxis_title='Total Score',template="plotly_dark")
water.show()
```

# **3D MESH Plot**

```python
import plotly.graph_objects as go
import pandas as pd
# Load the data from a CSV file
df = pd.read_csv("/content/AI_index_db (1).csv")
# Data for the mesh plot
data = [
    go.Mesh3d(
        x=df['Talent'],
        y=df['Infrastructure'],
        z=df['Operating Environment'],
        opacity=0.5,
        color='rgba(244,22,100,0.6)'
    )
]

# Layout settings for the plot
layout = go.Layout(
    title='3D Mesh Plot of Talent, Infrastructure, and Operating Environment',
    scene=dict(
        xaxis=dict(title='Talent'),
        yaxis=dict(title='Infrastructure'),
        zaxis=dict(title='Operating Environment')
    )
)

# Create the figure
fig1 = go.Figure(data=data, layout=layout)

# Display the plot
fig1.show()
```
